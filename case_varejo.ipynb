```python
# IMPORTANDO AS BILIOTECAS QUE SERÃO UTILIZADAS
import pandas as pd
import seaborn as sns
```

```python
#IMPORTANDO AS BASES DE DADOS
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda')['idcompra'].nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda')['idcompra'].agg('nunique')
```

```python
vendas.groupby('bandeira')['idcompra'].agg('count')
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento')['idcompra'].agg('count')
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas['estado'] = vendas['estado'].fillna('MS')
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'] = vendas['Preço'].fillna(media_preco)
vendas.head()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
vendas_correto = vendas.query("Preço < Preço_com_frete")
vendas_correto.describe()
```

```python
vendas_correto.query('Nome_Departamento == "Esporte_e_Lazer" and estado =="SP"').describe()
```

```python
#Departamentos mais vendidos

Dep_mais_vendidos = vendas_correto.groupby('Nome_Departamento')['idcompra'].agg('count').sort_values(ascending=False).head(5).reset_index()
Dep_mais_vendidos


```

```python
#media de preço com frete por departamen
Dep_media_preco = vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').reset_index().sort_values(by='Preço_com_frete',ascending=False).head(5)
Dep_media_preco
```

```python
#quantidade de vendas por mes

vendas_correto.loc[:,'mes'] = vendas_correto['Data'].dt.month

vendas_correto.groupby('mes')['idcompra'].agg('nunique').sort_values(ascending=False).head(10)
```

```python
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
cliente[['idade','renda']].describe()
```

```python
cliente.info( )
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
# juntando a tabela de vendas com a tabela de clientes

vendas_cliente = vendas_correto.merge(cliente, how='left', on='cliente_Log')
vendas_cliente.head()
```

```python
# Renda média dos clientes para cada tipo de canal de vendas
agg_renda_canal = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_renda_canal
```

```python
#Idade média dos clientes por bandeira
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(4,2))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='green')
plt.ylabel("Média de Idade")
plt.title(label = "Idade média por Bandeira", loc="Center", fontsize=18, fontweight='bold', color='Blue')
```

```python
fig2 = plt.figure(figsize=(4,2))
plt.bar(agg_renda_canal['idcanalvenda'], agg_renda_canal['renda'], color='brown')
plt.ylabel("Renda Média")
plt.title("Renda Média por canal de venda", loc="center", fontsize = 12, fontweight='bold', color= 'blue')
```

```python
venda_por_data = vendas_correto.groupby('Data')['idcompra'].agg('nunique').reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(7,3))
plt.xlabel("Data da venda")
plt.ylabel("Quantidade de Vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]
plt.plot(x,y, color="blue", linewidth="2")
plt.title('Quantidade de Vendas')
```

```python
# Média de preço com frete por Nome de Departamento.
media_preco_dep = round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)

fig = plt.figure(figsize=(25, 7))
plt.xlabel('Departamento')
plt.ylabel('Média de Preço com Frete')
x = media_preco_dep['Nome_Departamento']
y = media_preco_dep['Preço_com_frete']
plt.bar(x, y)
plt.title("Preço médio por departamento", loc='Center', size=20, fontweight='bold', color='#625696')

plt.show()

```
